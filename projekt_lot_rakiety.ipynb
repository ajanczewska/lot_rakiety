## Definicje i&nbsp;wzory

<b>Rakieta</b> - obiekt latajƒÖcy wprawiany w&nbsp;ruch silnikiem odrzutowym, s≈Çu≈ºƒÖcy do lot√≥w kosmicznych.<a href='sources'>$^{[4]}$</a>

Je≈ºeli ruch rakiety odbywa siƒô w&nbsp;przestrzeni kosmicznej, to si≈Çy zewnƒôtrzne ($F_{zew}$) sƒÖ zaniedbywane i&nbsp;wtedy zmiana pƒôdu rakiety jest r√≥wna sile ciƒÖgu (jest spe≈Çniona zasada zachowania pƒôdu). Natomiast gdy ruch odbywa siƒô w&nbsp;pobli≈ºu Ziemi (np. tu≈º po starcie), to w√≥wczas $F_{zew}$ reprezentuje ciƒô≈ºar rakiety oraz si≈Çƒô oporu atmosfery i&nbsp;trzeba je uwzglƒôdniƒá. Konstruktorzy rakiet starajƒÖ siƒô uzyskaƒá jak najwiƒôkszƒÖ si≈Çƒô ciƒÖgu, aby przezwyciƒô≈ºyƒá $F_{zew}$.<a href='sources'>$^{[5]}$</a>

<b>Rakieta wielostopniowa, rakieta wielocz≈Çonowa</b> - rakieta podzielona na kilka cz≈Çon√≥w (stopni) napƒôdzajƒÖcych jƒÖ kolejno.
Ka≈ºdy cz≈Çon ma w≈Çasny silnik i&nbsp;materia≈Çy pƒôdne. Ostatni stopie≈Ñ zawiera statek kosmiczny lub (w&nbsp;przypadku pocisku rakietowego) g≈Çowicƒô z&nbsp;≈Çadunkiem bojowym.<a href='sources'>$^{[6]}$</a>

Zalety rakiet wielostopniowych<a href='sources'>$^{[7]}$</a>:
* mo≈ºliwo≈õƒá skorzystania z&nbsp;innego typu silnika - lepiej dopasowanego do warunk√≥w jego pracy - dla ka≈ºdego stopnia. W&nbsp;‚Äûdolnych‚Äù cz≈Çonach stosuje siƒô silniki dostosowane do ci≈õnienia atmosferycznego, w&nbsp;‚Äûg√≥rnych‚Äù natomiast - silniki zaprojektowane do dzia≈Çania w&nbsp;warunkach bliskich pr√≥≈ºni;
* pozbywanie siƒô zbƒôdnej masy w&nbsp;trakcie lotu;
* wed≈Çug Konstantina Cio≈Çkowskiego rakieta jednostopniowa, ze wzglƒôdu na zbyt du≈ºy ciƒô≈ºar, nie jest w&nbsp;stanie osiƒÖgnƒÖƒá orbity ziemskiej.

Wady rakiet wielostopniowych<a href='sources'>$^{[7]}$</a>:
* zwiƒôkszenie stopnia skomplikowania konstrukcji, co skutkuje wiƒôkszym zagro≈ºenie awariami - nieprawid≈Çowe oddzielenie kt√≥rego≈õ ze stopni, kolizje poszczeg√≥lnych cz≈Çon√≥w bƒÖd≈∫ niepoprawny zap≈Çon kolejnego silnika;
* konieczno≈õƒá przenoszenia dodatkowego ciƒô≈ºaru w&nbsp;postaci silnik√≥w niewykorzystywanych w&nbsp;wiƒôkszo≈õci faz lotu. 

### Przyspieszenie rakiety z&nbsp;uwglƒôdnieniem si≈Çy grawitacji<a href='sources'>$^{[5]}$</a><a href='sources'>$^{[8]}$</a>

Rozwa≈ºmy rakietƒô, na kt√≥rƒÖ dzia≈Ça si≈Ça grawiatcji, skierowana pionowo w&nbsp;d√≥≈Ç. Od momentu startu ($t = 0$) rakieta wyrzuca gaz przy sta≈Çym strumieniu masy $R = \frac{dm}{dt}$ (gdzie $dm$ - masa wyrzucanego gazu, a&nbsp;$dt$ - czas, w&nbsp;kt√≥rym gaz jest wyrzucany). Prƒôdko≈õƒá spalin wzglƒôdem rakiety jest r√≥wna $v_{e}$.

Przyspiesznie rakiety wynosi wtedy:
$$a = \frac{v_{e}}{m}R - g = \frac{v_{e}}{m}\frac{dm}{dt} -g.$$

Poszczeg√≥lne elementy we wzorze majƒÖ nastƒôpujƒÖce znaczenie:
* im wiƒôksza jest warto≈õƒá prƒôdko≈õci wyrzucanego paliwa wzglƒôdem rakiety ($v_{e}$), tym wiƒôksze jest przyspieszenie rakiety. Granica $v_{e}$ dla konwencjonalnych uk≈Çad√≥w napƒôdowych na gorƒÖcy gaz wynosi oko≈Ço $2,5 \cdot 10^3\frac{m}{s}$;
* iloczyn $(\frac{dm}{dt})v_{e}$ to si≈Ça ciƒÖgu rakiety. Im szybciej rakieta spala paliwo, tym wiƒôksza jest ta si≈Ça, z&nbsp;czego wynika, ≈ºe przyspieszenie rakiety jest wiƒôksze;
* masa rakiety $m$ drastycznie spada podczas lotu, poniewa≈º jej wiƒôkszo≈õƒá to paliwo.
Przyspieszenie ro≈õnie w&nbsp;spos√≥b ciƒÖg≈Çy, osiƒÖgajƒÖc maksimum tu≈º przed wyczerpaniem siƒô paliwa.

Prƒôdko≈õƒá ko≈ÑcowƒÖ rakiety, a&nbsp;tak≈ºe prƒôdko≈õƒá w&nbsp;ka≈ºdym momencie jej lotu, mo≈ºemy opisaƒá za pomocƒÖ r√≥wnania Cio≈Çkowskiego.

### R√≥wnanie Cio≈Çkowskiego<a href='sources'>$^{[5]}$</a>
<b>R√≥wnanie rakiety Cio≈Çkowskiego</b> ‚Äì r√≥wnanie matematyczne, opisujƒÖce ruchy pojazd√≥w, zachowujƒÖcych siƒô tak jak rakiety. SƒÖ to urzƒÖdzenia, kt√≥re mogƒÖ nadawaƒá sobie przyspieszenie za pomocƒÖ si≈Çy ciƒÖgu poprzez pozbycie siƒô czƒô≈õci swojej masy z&nbsp;du≈ºƒÖ prƒôko≈õciƒÖ i&nbsp;w&nbsp;ten spos√≥b zachowujƒÖc pƒôd.

\begin{equation}
    \Delta v = v_{e} \ln{\frac{m_{0}}{m_{1}}},
\end{equation}

gdzie:
* $\Delta v$ - ko≈Ñcowa (maksymalna) prƒôdko≈õƒá pojazdu;
* $m_{0}$ - ca≈Çkowita masa poczƒÖtkowa pojazdu wraz z&nbsp;paliwem;
* $m_{1}$ - masa ko≈Ñcowa pojazdu bez paliwa;
* $v_{e}$ - prƒôdko≈õƒá wydzielanego czynnika roboczego (spalonego paliwa) wzglƒôdem pojazdu.

### Przyk≈Çad<a href='sources'>$^{[8]}$</a>:
Obliczmy stosunek mas potrzebny do ucieczki spod oddzia≈Çywania grawitacji ziemskiej, zaczynajƒÖc od momentu, kiedy rakieta spoczywa i&nbsp;zak≈ÇadajƒÖc, ≈ºe prƒôdko≈õƒá ucieczki z&nbsp;Ziemi wynosi oko≈Ço $11,2 \cdot 10^3\frac{m}{s}$, a&nbsp;prƒôdko≈õƒá wyrzucania spalin wzglƒôdem rakiety to $v_{e}= 2,5  \cdot 10^3\frac{m}{s}$.

Przekszta≈Çcamy wz√≥r Cio≈Çkowskiego i&nbsp;otrzymujemy:
$$\frac{m_{0}}{m_{1}} = e^{\frac{v}{v_{e}}}.$$

Nastƒôpnie podstawiamy dane: 
$$\frac{m_{0}}{m_{1}} = e^{\frac{11,2 \cdot 10^3\frac{m}{s}}{2,5 \cdot 10^3\frac{m}{s}}} = e^{4,48} = 88,$$
$$m_{1} = \frac{m_{0}}{88}.$$

<strong>Wynika z&nbsp;tego, ≈ºe po spaleniu paliwa pozostaje jedynie $\frac{1}{88}$ masy poczƒÖtkowej, czyli paliwo stanowi, a≈º $\frac{87}{88}$ masy poczƒÖtkowej rakiety.</strong>

### Wyprowadzenie wzoru Cio≈Çkowskiego<a href='sources'>$^{[9]}$</a>

Rysunek poglƒÖdowy
<center><img src="myszojelen_w_rakiecie.png" title="Myszojele≈Ñ w rakiecie" /></center>

Wiemy, ≈ºe prƒôdko≈õƒá spalin wzglƒôdem rakiety ($v_{e}$) dana jest zale≈ºno≈õciƒÖ:
$$v_{e} = v - v_{s},$$
gdzie:
* $v_{s}$ - prƒôdko≈õƒá spalin opuszczajƒÖcych silnik wzglƒôdem obserwatora (na Ziemi);
* $v$ - prƒôdko≈õƒá chwilowa rakiety wzglƒôdem obserwatora (na Ziemi).
<br><br>

Je≈ºeli w&nbsp;przedziale czasu $dt$ z&nbsp;rakiety zostaje wyrzucona masa $dm_{s}$ z&nbsp;prƒôdko≈õciƒÖ $v_{s}$ to masa rakiety maleje o&nbsp;$dm$, a&nbsp;jej prƒôdko≈õƒá ro≈õnie o&nbsp;$\Delta v$. Przy czym:
$$\frac{dm_{s}}{dt} = -\frac{dm}{dt},$$
ze znakiem minus, poniewa≈º rakieta traci swojƒÖ masƒô.

Zmiana pƒôdu uk≈Çadu ($dP$) w&nbsp;czasie $dt$ wynosi:
\begin{equation}
    \frac{dP}{dt} = \frac{dP_{rakiety}}{dt} + \frac{dP_{spalin}}{dt} = \frac{d(mv)}{dt} + v_{s}\frac{dm_{s}}{dt} = m\frac{dv}{dt} + v\frac{dm}{dt} + v_{s}\frac{dm_{s}}{dt},
\end{equation}
zak≈ÇadajƒÖc, ≈ºe prƒôdko≈õƒá wyrzucania spalin jest sta≈Ça.

Wiemy, ≈ºe:
$$-v_{e}\frac{dm_{s}}{dt} = v\frac{dm}{dt} + v_{s}\frac{dm_{s}}{dt}.$$

Zgodnie z&nbsp;drugƒÖ zasadƒÖ dynamiki Newtona:
\begin{equation}
     F_{zew} = \frac{dP}{dt} = m\frac{dv}{dt} - v_{e}\frac{dm_{s}}{dt}.
\end{equation}

Ostatni wyraz w&nbsp;r√≥wnaniu mo≈ºe byƒá interpretowany jako si≈Ça wywierana na uk≈Çad przez substancjƒô (spaliny), kt√≥ra z&nbsp;niego wylatuje. W&nbsp;przypadku rakiety (samolotu) nosi ona nazwƒô si≈Çy ciƒÖgu.

Je≈ºeli na uk≈Çad nie dzia≈ÇajƒÖ ≈ºadne zewnƒôtrzne si≈Çy, wtedy $F_{zew} = 0$.

Wynika z&nbsp;tego, ≈ºe:
$$m\frac{dv}{dt} = v_{e}\frac{dm_{s}}{dt}.$$

Po podstawieniu $dm_{s} = -dm$ i&nbsp;przekszta≈Çceniu r√≥wnania otrzymujemy:
$$dv = -v_{e}\frac{dm}{m}.$$

Zak≈ÇadajƒÖc, ≈ºe $v_{e}$ jest sta≈Ça otrzymujemy:
\begin{equation}
    \int_{0}^{\Delta v} \,dv = -v_{e}\int_{m_{0}}^{m_{1}} \frac{1}{m} \, dm,
\end{equation}
gdzie:
* $\Delta v$ - prƒôdko≈õƒá ko≈Ñcowa rakiety;
* $m_{0}$ - masa poczƒÖtkowa rakiety z&nbsp;paliwem;
* $m_{1}$ - masa ko≈Ñcowa rakiety bez paliwa.

Ostatecznie otrzymujemy:
\begin{equation}
    \Delta v = v_{e}\ln{\frac{m_{0}}{m_{1}}}.
\end{equation}

## Wykresy zale≈ºno≈õci bez uwzglƒôdniania si≈Ç zewnƒôtrznych dzia≈ÇajƒÖcych na rakietƒô

<b>Wz√≥r na zmianƒô prƒôdko≈õci rakiety w&nbsp;czasie:</b>
$$v = v_e\ln\left( \frac{m}{m - \frac{dm}{dt}t} \right),$$
gdzie:
* $t$ - czas od poczƒÖtku startu rakiety;
* $m$ - masa poczƒÖtkowa rakiety z&nbsp;paliwem.

Drogƒô pokonanƒÖ przez rakietƒô obliczamy ze wzoru $s = v\cdot t$, podstawiajƒÖc $v$&nbsp;obliczone w&nbsp;ka≈ºdej sekundzie jej lotu, wiƒôc $t$ zawsze r√≥wna siƒô $1$s.

Przyspieszenie rakiety obliczamy ze wzoru $a = \frac{\Delta v}{\Delta t}$ dla $\Delta t = 1$s.

```python
import math
import matplotlib.pyplot as plt
import numpy as np

def distance(v):
    """Count distance travelled by rocket in every second.

    Args:
        v [list]: rocket velocity in every second.

    Return:
        [list]: distance travelled by rocket in every second.
    """
    s = [0]
    current_s = 0
    for current_velocity in v[0:len(v)-1]:
        current_s += current_velocity
        s.append(current_s/1000)
    return s

def velocity(i, M, ve, f):
    """Count rocket velocity.

    Args:
        i [int]: time from launching rocket (in seconds),
        M [int]: mass of rocket with propellant,
        ve [int]: exhaust velocity,
        f [int]: mass of ejected gas per second.

    Return:
        [int]: rocket velocity in i.
    """
    v = ve * math.log(M/(M-f*i))
    return v

def acceleration(v):
    """Count rocket acceleration.

    Args:
        v [list]: rocket velocity in every second.

    Return:
        [list]: rocket acceleration in every second.
    """
    a = []
    for i in range(len(v)-1):
        a.append((v[i+1]-v[i])/1)
    return a

def plot1(x, M, ve, f, variable, var_values, number):
    """Draw velocity vs. time plot, acceleration vs. time plot and distance vs. time plot.

    Args:
        x [numpy.linspace]: time interval (in seconds),
        M [list]: masses of rocket with propellant,
        ve [list]: exhaust velocities,
        f [list]: masses of ejected gas per second.
    """
    fig, (ax0, ax1, ax2) = plt.subplots(nrows=1, ncols=3, sharex=True, figsize=(19, 5.5))

    v0 = [velocity(i, M[0], ve[0], f[0]) for i in x]
    v1 = [velocity(i, M[1], ve[1], f[1]) for i in x]
    v2 = [velocity(i, M[2], ve[2], f[2]) for i in x]
    v3 = [velocity(i, M[3], ve[3], f[3]) for i in x]

    ax0.set_title('{}.1.Zale≈ºno≈õƒá drogi pokonanej przez rakietƒô od czasu lotu'.format(number), fontsize=13)
    for list_of_velocities in [v0, v1, v2, v3]:
        ax0.plot(x, distance(list_of_velocities))
    ax0.legend([variable + ": " + str(v) for v in var_values])
    ax0.set_ylim(0)
    ax0.set_xlabel('czas [s]', fontsize=10)
    ax0.set_ylabel('droga [km]', fontsize=10)

    ax1.set_title('{}.2.Zale≈ºno≈õƒá prƒôdko≈õci rakiety od czasu lotu'.format(number), fontsize=13)
    for list_of_velocities in [v0,v1,v2,v3]:
        ax1.plot(x, list_of_velocities)
    ax1.legend([variable + ": " + str(v) for v in var_values])
    ax1.set_ylim(0)
    ax1.set_xlabel('czas [s]', fontsize=10)
    ax1.set_ylabel('prƒôdko≈õƒá [m/s]', fontsize=10)

    ax2.set_title('{}.3.Zale≈ºno≈õƒá przyspieszenia rakiety od czasu lotu'.format(number), fontsize=13)
    for list_of_velocities in [v0,v1,v2,v3]:
        ax2.plot(x[1:len(x)], acceleration(list_of_velocities))
    ax1.legend([variable + ": " + str(v) for v in var_values])
    ax2.legend([variable + ": " + str(v) for v in var_values])
    ax2.set_ylim(0)
    ax2.set_xlabel('czas [s]', fontsize=10)
    ax2.set_ylabel('przyspieszenie [m/s¬≤]', fontsize=10)

    fig.suptitle('Wykresy zale≈ºno≈õci, zmienna - {}'.format(variable), fontsize=16)
    plt.show()

if __name__ == "__main__":
    x = np.linspace(0, 1200, 1201)
    plot1(x, M1, ve1, f1, data[0][1], M1, 1)
    plot1(x, M2, ve2, f2, data[0][2], ve2, 2)
    plot1(x, M3, ve3, f3, data[0][3], f3, 3)
```

## R√≥wnanie rakiety jednostopniowej z&nbsp;uwzglƒôdnieniem si≈Çy grawitacji i&nbsp;oporu atmosferycznego

R√≥wnanie opisujƒÖce si≈Çƒô wypadkowƒÖ dzia≈ÇajƒÖcƒÖ na rakietƒô, przy uwzglƒôdnienieniu, ≈ºe:
$$F = F_c - mg - kv^2,$$
gdzie:
* $F = \frac{d(mv)}{dt}$ - si≈Ça wypadkowa;
* $F_c = const$ - si≈Ça ciƒÖgu rakiety;
* $g = const$ - przyspieszenie grawitacyjne;
* $m$ - masa rakiety zmieniajƒÖca siƒô w&nbsp;czasie;
* $k$ - wsp√≥≈Çczynnik oporu atmosferycznego;
* $v$ - prƒôdko≈õƒá rakiety zmieniajƒÖca siƒô w&nbsp;czasie.

Powy≈ºsze r√≥wnanie mo≈ºemy r√≥wnie≈º zapisaƒá w&nbsp;nastƒôpujƒÖcej formie:
$$m\frac{dv}{dt} + \frac{dm}{dt}v = F_{c} - mg - kv^2.$$

Niech masa rakiety bƒôdzie r√≥wna:
$$m = m_0 - \gamma t,$$
gdzie:
* $m_0$ - masa poczƒÖtkowa rakiety (masa rakiety z&nbsp;paliwem);
* $\gamma = \frac{dm}{dt}$ - zmiana masy rakiety w&nbsp;czasie, to znaczy warto≈õƒá tempa spalania paliwa;
* $t$ - czas lotu rakiety.

Podstawiamy $m$ oraz $\frac{dm}{dt}$ do powy≈ºszego r√≥wnania i&nbsp;otrzymujemy:
$$(m_0 - \gamma t)\frac{dv}{dt} -\gamma v = F_c - (m_{0} -\gamma t)g - kv^2.$$

### Metoda Eulera

Aby znale≈ºƒá numeryczne rozwiƒÖzanie r√≥wnania: $(m_0 - \gamma t)\frac{dv}{dt} -\gamma v = F_c - (m_{0} -\gamma t)g - kv^2$, korzystamy z&nbsp;metody Eulera.

Najpierw przekszta≈Çcamy r√≥wnanie do formy:
$$\frac{dv}{dt} = \frac{F_c - (m_{0} -\gamma t)g - kv^2 + \gamma v}{m_0 - \gamma t}.$$

<strong>Stosujemy metodƒô Eulera do naszego r√≥wnania.</strong>

PochodnƒÖ przybli≈ºamy r√≥≈ºnicƒÖ sko≈ÑczonƒÖ:
$$ v'(t_n) \approx \frac{v(t_n + \Delta t)- v(t_n)}{\Delta t} ,$$
gdzie:
* $\Delta t$ - przyrost czasu,

przy warunku poczƒÖtkowym: $v(0) = v_0$.

Niech $t_n + \Delta t = t_{n+ 1}$.

Je≈ºeli $v_n$ oznacza numeryczne przybli≈ºenie do $v(t_n)$ otrzymujemy:
$$v_{n + 1} = v_n + \Delta t \frac{dv}{dt}.$$

Podstawimay $\frac{dv}{dt}$ i&nbsp;otrzymujemy:
$$v_{n + 1} = v_n + \frac{F_c - (m_{0} -\gamma t_n)g - kv_n^2 + \gamma v_n}{m_0 - \gamma t_n}\Delta t.$$

<strong>U≈ºyjemy wzoru, aby przybli≈ºyƒá numeryczne rozwiƒÖzanie r√≥wnania dla konkretnej rakiety.</strong>

W≈Ça≈õciwo≈õci rakiety:
* $F_c = 34000000$ N;

* $m_0 = 2500000$ kg;

* $\gamma = 1000 \frac{kg}{s}$;

* $g = 9,8 \frac{m}{s^2}$;

* $k = 0.4 \frac{kg}{m}$;

Niech $\Delta t = 1 s$.

Warunek poczƒÖtkowy: $v(0) = 0$.

Podstawimy dane do r√≥wnania:
$$v_{n + 1} = v_n + \frac{34000000 - (2500000 - 1000t_n)9,8 - 0.4v_n^2 + 1000v_n}{2500000 - 1000t_n} = \frac{9500000 + 98000t_n - 0.4v_n^2 + 1000v_n}{2500000 - 1000t_n }.$$

### Wykres pokazujƒÖcy dok≈Çadno≈õƒá metody numerycznej Eulera

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [19, 5.5]
from scipy.integrate import odeint
import numpy as np

def euler_velocity(v_prev, max_time):
    """Count rocket velocity using Euler's methode.

    Args:
        v_prev [int]: initial rocket velocity,
        max_time [int]: maximum time value.

    Return:
        [list]: list of rocket velocities changing every second.
    """
    velocities = [v_prev]
    for t in range(1, max_time):
        v_next = v_prev + ((9500000) + 98000*t - 0.4*(v_prev**2) + 1000*v_prev)/(2500000 - 1000*t)
        velocities.append(v_next)
        v_prev = v_next
    return velocities

def diff_velocity(v, t):
    """Count diffrential value of v at t time."""
    dvdt = (9500000 + 98000*t - 0.4*(v**2) + 1000*t)/(2500000 - 1000*t)
    return dvdt

def plot3(time_range, solve, euler):
    """Draw velocity vs. time plot.

    Args:
        time_range [numpy.linspace]: time interval (in seconds),
        solve [list]:  list of velocities for diffrential equation,
        euler [list]: list of velocities for Euler's methode.
    """
    plt.plot(time_range, solve)
    plt.plot(time_range, euler)
    plt.title("5.Wykres zale≈ºno≈õci prƒôdko≈õci rakiety od czasu lotu", fontsize=16)
    plt.xlabel("czas [s]", fontsize=12)
    plt.ylabel("prƒôdko≈õƒá [m/s]", fontsize=12)
    plt.legend(["Funkcja Odeint wbudowana w Pythonie", "Metoda Eulera"], fontsize=12)
    plt.show()

if __name__=="__main__":
    v0 = 0.0
    max_time = 600
    time_range = np.linspace(0, 600, 600)

    solve = odeint(diff_velocity, v0, time_range)
    euler = euler_velocity(v0, max_time)
    plot3(time_range, solve, euler)
    
```

### Analiza wykresu
Wykres pokazuje, ≈ºe metoda Eulera daje dok≈Çadne przybli≈ºenia prƒôdko≈õci rakiety dla czasu lotu od 0 do 3 minut. W&nbsp;p√≥≈∫niejszym przedziale czasu przybli≈ºanie prƒôdko≈õci tƒÖ metodƒÖ jest coraz mniej dok≈Çadne.

### Wykres zale≈ºno≈õci prƒôdko≈õci od czasu w&nbsp;zale≈ºno≈õci od si≈Ç dzia≈ÇajƒÖcych na rakietƒô

```python
import math
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [19, 5.5]
import numpy as np

def vel(v_init, max_time):
    """Count rocket velocity.

    Args:
        v_init [int]: initial rocket velocity,
        max_time [int]: maximum time value.

    Return:
        [list]: list of rocket velocities changing every second.
    """
    velocities = [v_init]
    for t in range(1, max_time):
        v = 34000 * math.log(2500000/(2500000 - 1000*t))
        velocities.append(v)
    return velocities

def plot4(time_range, velocities_tab, max_time):
    """Draw velocity vs. time plot.

    Args:
        velocity_tab [list]: list of rocket velocities,
        time_range [numpy.linspace]: time interval (in seconds),
        max_time [int]: maximum time value.
    """
    plt.plot(time_range, velocities_tab[0])
    plt.plot(time_range, velocities_tab[1])

    plt.title("6.Zale≈ºno≈õƒá prƒôdko≈õci rakiety od czasu lotu", fontsize=16)
    plt.legend(['warto≈õƒá prƒôdko≈õci, gdy nie dzia≈ÇajƒÖ zewnƒôtrzne si≈Çy', 'warto≈õc prƒôdko≈õci, gdy dzia≈Ça si≈Ça grawitacji i oporu atmosferycznego'], fontsize=12)
    plt.xlim(0, max_time)
    plt.xlabel('czas [s]', fontsize=12)
    plt.ylabel('prƒôdko≈õƒá rakiety [m/s]', fontsize=12)
    plt.show()

if __name__=="__main__":
    time_range = np.linspace(0, 180, 180)
    v0 = 0
    max_time = 180
    plot4(time_range, [vel(v0, max_time), euler_velocity(v0, max_time)], max_time)
```
